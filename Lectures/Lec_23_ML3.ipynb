## 23.1 Load the data

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
from sklearn import decomposition
```

## 23.2 Choose the estimator

```python
from sklearn import decomposition
decomposition
pca = decomposition.PCA(n_components=2)
pca.fit(X)
X2 = pca.transform(X)
print(X2)
```

## 23.3 Visualize the data

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(1, figsize=(8, 6))
#fig = plt.figure()

labels = ['Setosa', 'Versicolour', 'Virginica']
values = [0, 1, 2]
#for index in range(len(labels)):
#    value = values[index]
#    label = labels[index]
#for i, label in enumerate(labels):
#    print(i, label)
#for i in range(len(labels)):
#    print(i, labels[i])

    
#for i in range(len(labels)):
#    print(values[i], labels[i])
    
#for value, label in zip(values, labels):
#    print(value, label)

for value, label in zip(values, labels):
    data = X2[y==value]
    #print(data)
    plt.scatter(data[:,0], data[:,1], label=label)
plt.legend()
plt.show()
```

```python
from sklearn import decomposition

pca = decomposition.PCA(n_components=3)
pca.fit(X)
X3 = pca.transform(X)
print(X3)
```

## 23.3 Different decomposition

```python
#http://scikit-learn.org/stable/auto_examples/decomposition/plot_pca_iris.html
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(1, figsize=(8, 6))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
plt.cla()

for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(X3[y == label, 0].mean(),
              X3[y == label, 1].mean() + 1.5,
              X3[y == label, 2].mean(), name,
              #horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
y1 = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y1, #cmap=plt.cm.spectral,
           edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
