```python
import numpy as np
from matplotlib import pyplot as plt


def Mandelbrot(size, iters):
    
    
    a1, b1, points1 = -2, 2, size  # Side of x
    a2, b2, points2 = -2, 2, size  # Side of y
    N_iter = iters                 # the maximum number of iterations

    x_set = np.linspace(a1, b1, points1)
    y_set = np.linspace(a2, b2, points2)

     
    mandset = np.zeros([points1, points2], float) # Empty Array

    # Calculate the values in the array
    for i in range(len(y_set)): 
        for j in range(len(x_set)):

            # initial conditions
            z = 0        
            c = x_set[j] + y_set[i]*1j       
            inset = True

            # need to iterate it to see if the point satisfies the condition
            for k in range(N_iter):
                z = z**2 + c
                # print('values:  ', k,z,c)
                if np.absolute(z) > 2:
                    mandset[i, j] = k
                    inset = False
                    break
    
    # Make the plot
    title = "(" + str(size) +  'x' + str(size) + ', ' + str(iters) + ' iterations)'
    fig = plt.figure(figsize=(2, 2), dpi = 600)
    ax = fig.add_axes([0.0, 0.0, 1.0, 1.0], frameon=False, aspect=1)
    plt.title(title)
    plt.imshow(mandset, origin="lower", extent=[a1,b1, a2,b2], cmap="coolwarm")
    plt.show()
   
```

```python
def Mandelbrot2(size, iters):
    """
    Generate and plot a Mandelbrot set with square grid size, "size" x "size,"
    and finite number of iterations, "iters."
    """
    
    
    a1, b1, points1 = -2, 2, size  # Side of x
    a2, b2, points2 = -2, 2, size  # Side of y
    N_iter = iters                 # the maximum number of iterations

    x_set = np.linspace(a1, b1, points1)
    y_set = np.linspace(a2, b2, points2)
    data = []

     
    mandset = np.zeros([points1, points2], float) # Empty Array

    # Calculate the values in the array
    for i in range(len(y_set)): 
        for j in range(len(x_set)):

            # initial conditions
        
            z = 0        
            c = x_set[j] + y_set[i]*1j       
            inset = True

            # need to iterate it to see if the point satisfies the condition
            for k in range(N_iter):
                z = z**2 + c
                #print('values:  ', k,z,c)
                if np.absolute(z) > 2:
                    mandset[i, j] = k
                    inset = False
                    break
        
            data.append(z)
    #print (data)
    #print(len(data))
    z1 = np.fft.fft(data)
    plt.title("FFT of 1D Mandelbrot Set")
    plt.plot(np.abs(z1))
    plt.xlabel("Frequency Sampling(Fs)")
    plt.ylabel('Amplitude')
    plt.show()
    
    z2 = np.fft.ifft(z1)
    plt.title("IFFT of FFT 1D Mandelbrot Set")
    plt.plot(np.abs(z2))
    plt.xlabel("Frequency")
    plt.ylabel('Amplitude')
    plt.show()
    
    data2 = data
    data2 = np.mgrid[:size, :size][0]
    #np.reshape(data, (size, size))
    #np.array(data2).reshape(size,size)
    z3 = np.fft.fft2(data2)
    plt.title("FFT of Mandelbrot set")
    plt.plot(np.abs(z3))
    plt.xlabel("Fs")
    plt.ylabel('Amplitude')
    plt.show()
    
    ## FFT of 2D array
    z4 = np.fft.ifft2(z3).real
    plt.figure(figsize=[10, 5])
    plt.title("FFT Reversed back ")
    plt.imshow(z4)
    
    
    
    
        
    
```
