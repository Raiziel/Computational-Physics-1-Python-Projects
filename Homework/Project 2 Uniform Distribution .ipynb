```python
# A code to generate pseudo random numbers
import numpy as np
from scipy.special import entr 
## here I'm importing the entropy function to calculate the entropy value that is passed in from my data set
import matplotlib.pyplot as plt

N = 100
a = 1624
c = 1012
m = 11 #2**31-1 #4245035300
x = 1
results = []
t = 100
time=[]
for i in range (t):
    time.append(i)
for i in range(N):      
    x = (a*x+c)%m
    results.append(x)
#print (len(results))
#data = results
#print (results)
#print (entr(results))
plt.title("Distribution of Pseuo-Random Numbers Generated")
plt.plot(results,"o")
plt.show()
data = entr(np.array(results)/sum(np.array(results)))
print (data)
plt.title("Distribution of Entropy Values of PRNG")
plt.ylabel("Occurence of Random Elements")
plt.plot(data,"o")
plt.show()
```
