```python
## Program written by Nicholas V. Munoz
f = lambda x: x**2 - x + 1 ## range 0 - 3 
def simpson(start,end,parts): ## simpson's rule h/3(f(start)+ 4*f(odds) + 2*f(evens) + f(end))
    h = ((end-start)/parts) 
    
    Area = (h/3)*(f(start)+ 4*f(((start+end)/2))+ f(end))
    
    print ("Simpson's Area:", Area)
    
```

```python
simpson(0,3,5)
```

```python
import math
f = lambda x: math.exp(-x**2) ## range 0 - 3
def trapezoid(start,end,parts):

    y0 = f(start)
    y1 = f(end)
    A = 0.5*((end-start)/parts)*(y1 + y0)
    print("Trapezoid area:",A)
```
