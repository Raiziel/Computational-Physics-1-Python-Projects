```python
##dowoloaded the data from http://www-personal.umich.edu/~mejn/cp/data/circular.txt


import matplotlib.pyplot as plt
import numpy as np
from cmath import exp, pi

data = np.loadtxt("circular.txt",float)
X = data
N = len(X)
def dft(X):
    '''
    This is a code to do discrete Fourier transform
    '''    
    N = len(X)
    c = np.zeros(N, complex)
    for k in range(N):
        for n in range(N):
            c[k] += X[n]*exp(-2j*pi*k*n/N)
    return c
plt.title("Fourier Transformation Set")
plt.plot(c)
plt.xlabel("x")
plt.ylabel("F(x)")
```

```python
plt.title("Inverse Fourier Transformation Set")
plt.plot(np.linalg.inv(c))
plt.xlabel("x")
plt.ylabel("F(x)")
```

```python
def dft1(X):
    '''
    This is DFT
    '''    
    N = len(X)
    c = np.zeros(N, complex)
    for k in range(N):
        for n in range(N):
            c[k] += X[n]*exp(-2j*pi*k*n/N)
    return c
def dft2(y):
    '''
    This is FFT
    '''
    x = np.asarray(y, dtype=float)        #N*1 array
    N = y.shape[0]
    n = np.arange(N)                      #1*N array
    k = n.reshape((N, 1))                 #N*1 array
    M = np.exp(-2j * np.pi * k * n / N)   #N*N matrix
    return np.dot(M, x)                   #N*1 array
```
