```python
## Program Written by Nicholas Munoz
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=[10, 6])


#define the function
f = lambda x,a,b,c: a*x**2 + 6*b*x +c  

#define the paramters for the plot
x_min, x_max = 0, 10
npoints = 10
a,b,c = 1, 2, 1

x = np.linspace(x_min, x_max, npoints)
y = f(x,a,b,c) + np.random.rand(npoints) - 0.5

#do the curve fit
params, extras = curve_fit(f, x, y)

plt.plot(x,f(x,a,b,c), '--', label='ideal function')
plt.plot(x,y, 'o', label='noisy data')
plt.plot(x,f(x,params[0],params[1],params[2]), label='curve_fit')

plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.xlim([x_min, x_max])
plt.legend(fontsize=12)
plt.show()
print('original coefficients: %6.3f, %6.3f, %6.3f' %(a,b,c))
print('fitted   coefficients: %6.3f, %6.3f, %6.3f' %(params[0], params[1], params[2]))
```

```python
%matplotlib inline  

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=[12, 8])
#define the function
f = lambda x: x**2*np.cos(x+1)

#define the paramters for the plot
a,b = 0,3
npoints = 100
order = 10

x = np.linspace(a,b,npoints)
y = f(x) + np.random.rand(npoints) - ((b-a)/npoints)*order

fit = np.polyfit(x, y, order, full=True)
y_p = np.polyval(fit[a], x)

plt.plot(x,f(x), label='ideal function')
plt.plot(x,y, 'o', label='noisy data')
plt.plot(x,y_p, 'b', label='polyval fit'+str(order))
```

```python
## It seems that polynomials tend to have a pretty even distribution among the functions. Mostly, even functions 
## tend to align within reasonable fit of the function
## Unlike the sin function which has very large distances between points and less fitted coefficient distribution
```
