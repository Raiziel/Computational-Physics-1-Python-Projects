```python
## Decay process
import numpy as np
from scipy.special import entr 
import matplotlib.pyplot as plt


# Initial parameters
NTl = 1000
NPb = 0
tau = 3.053*60
dt = 1 
p = 1 - 2**(-dt/tau)
tmax = 1000

t_points = np.arange(0.0, tmax, dt)
Tl = []
Pb = []

for t in t_points:
    Tl.append(NTl)
    Pb.append(NPb)
    
    # Calculate the decayed atoms
    decay = 0
    for i in range(NTl):
        if np.random.random()<p:
            decay += 1
    NTl -= decay
    NPb += decay
plt.title("Radioactive Active Decay Tl to Pb")
plt.plot(t_points, Tl, label='Tl')
plt.plot(t_points, Pb, label='Pb')
plt.xlabel("Time")
plt.ylabel("Number of atoms")
plt.legend()
plt.show()
#print ((np.array(Tl)/sum(np.array(Tl))))

plt.xlabel("Time")
plt.ylabel("Entropy")
plt.title("Entropy over time with change in Particle Decay")
plt.plot(t_points,entr(((np.array(Tl)/sum(np.array(Tl))))))
plt.plot(t_points,entr(((np.array(Pb)/sum(np.array(Pb))))))

## This code first finds the probability of each individual element occuring within it's respective array
## Then finds the entropy value of that value occuring


```
