```python
##Written By Nicholas Munoz;downloaded the data from http://www-personal.umich.edu/~mejn/cp/data/circular.txt


import matplotlib.pyplot as plt
import numpy as np
from cmath import exp, pi

data = np.loadtxt("circular.txt",float)
X = data

plt.imshow(X)
plt.savefig("circular.png")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# plot of single line
plt.figure(figsize=[5, 5])
plt.title("Sliver of the data set")
plt.plot(data[1])
plt.xlabel("x")
plt.ylabel('Intensity (y)')
plt.show()
```

```python
## FFT of 2D array
Z = np.fft.fft2(X)
plt.title("Amplitude with Inverse FFT ")
plt.plot(np.abs(Z))
plt.xlabel("Inverse of X")
plt.ylabel('Intensity w/ Set Inverse X')
plt.show()
```

```python
## FFT of 2D array
Z2 = np.fft.ifft2(Z).real
plt.figure(figsize=[10, 5])
plt.title("FFT Reversed back ")
plt.imshow(Z2)
```

```python
## Real Part of Z ( not important information just testing graphs ) 
Z2 = np.fft.ifft2(Z.real)
plt.figure(figsize=[20, 15])
plt.title("IFFT Real")
plt.plot(Z2)
plt.xlabel("X")
plt.ylabel("y")
```

```python
def dft1(X):
    '''
    This is DFT
    '''    
    N = len(X)
    c = np.zeros(N, complex)
    for k in range(N):
        for n in range(N):
            c[k] += X[n]*exp(-2j*pi*k*n/N)
def dft2(y):
    '''
    This is FFT
    '''
    x = np.asarray(y, dtype=float)        #N*1 array
    N = y.shape[0]
    n = np.arange(N)                      #1*N array
    k = n.reshape((N, 1))                 #N*1 array
    M = np.exp(-2j * np.pi * k * n / N)   #N*N matrix
    return np.dot(M, x)                   #N*1 array
```
